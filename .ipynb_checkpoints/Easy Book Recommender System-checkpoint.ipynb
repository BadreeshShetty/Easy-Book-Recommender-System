## Recommendations based on correlations

```python
average_rating = pd.DataFrame(ratings.groupby('ISBN')['bookRating'].mean())
average_rating['ratingCount'] = pd.DataFrame(ratings.groupby('ISBN')['bookRating'].count())
average_rating.sort_values('ratingCount', ascending=False).head()
```

#### Observation: 

The book that received the most ratings is not highly rated at all.

#### To ensure statistical significance, users with less than 200 ratings, and books with less than 100 ratings are excluded.

```python
counts1 = ratings['userID'].value_counts()
ratings = ratings[ratings['userID'].isin(counts1[counts1 >= 200].index)]
counts = ratings['bookRating'].value_counts()
ratings = ratings[ratings['bookRating'].isin(counts[counts >= 100].index)]
```

### Rating matrix
The matrix will be sparse because not every user rate every book.

```python
ratings_pivot = ratings.pivot(index='userID', columns='ISBN').bookRating
userID = ratings_pivot.index
ISBN = ratings_pivot.columns
print(ratings_pivot.shape)
ratings_pivot.head()
```

Books correlated with the 3rd most rated book "The Da Vinci Code"

```python
bones_ratings = ratings_pivot['0385504209']
similar_to_bones = ratings_pivot.corrwith(bones_ratings)
corr_bones = pd.DataFrame(similar_to_bones, columns=['pearsonR'])
corr_bones.dropna(inplace=True)
corr_summary = corr_bones.join(average_rating['ratingCount'])
corr_summary[corr_summary['ratingCount']>=300].sort_values('pearsonR', ascending=False).head(10)
```

We obtained the books' ISBNs, but we need to find out the names of the books to see whether they make sense.

```python
books_corr_to_bones = pd.DataFrame(['044651652X', '044022165X', '006101351X', '0312278586', '0440222656', '0060987103', '0439064872', '0671021001', '0345370775'], 
                                  index=np.arange(9), columns=['ISBN'])
corr_books = pd.merge(books_corr_to_bones, books, on='ISBN')
corr_books
```

```python
combine_book_rating = pd.merge(ratings, books, on='ISBN')
columns = ['yearOfPublication', 'publisher', 'bookAuthor', 'imageUrlS', 'imageUrlM', 'imageUrlL']
combine_book_rating = combine_book_rating.drop(columns, axis=1)
combine_book_rating.head()
```

### Filter to only popular books

```python
combine_book_rating = combine_book_rating.dropna(axis = 0, subset = ['bookTitle'])
```

```python
book_ratingCount = (combine_book_rating.
     groupby(by = ['bookTitle'])['bookRating'].
     count().
     reset_index().
     rename(columns = {'bookRating': 'totalRatingCount'})
     [['bookTitle', 'totalRatingCount']]
    )
book_ratingCount.head()
```

#### Now we can merge the total rating count data into the rating data, giving us exactly what we need to filter out the lesser known books.

```python
rating_with_totalRatingCount = combine_book_rating.merge(book_ratingCount, left_on = 'bookTitle', right_on = 'bookTitle', how = 'left')
rating_with_totalRatingCount.head()
```

```python
pd.set_option('display.float_format', lambda x: '%.3f' % x)
print(book_ratingCount['totalRatingCount'].describe())
```

#### The median book has only been rated one time. Lets take a look at the top of the distribution.

```python
print(book_ratingCount['totalRatingCount'].quantile(np.arange(.9, 1, .01)))
```

#### So about 1% of books have 31 ratings, 2% have 19 ratings.

```python
popularity_threshold = 31
rating_popular_book = rating_with_totalRatingCount.query('totalRatingCount >= @popularity_threshold')
rating_popular_book.head()
```

#### Filtering to India,US and Canada users only

```python
combined = rating_popular_book.merge(users, left_on = 'userID', right_on = 'userID', how = 'left')
india_us_canada_user_rating = combined[combined['Location'].str.contains("india|usa|canada")]
india_us_canada_user_rating = india_us_canada_user_rating.drop('Age', axis=1)
india_us_canada_user_rating.head()
```

```python
if not india_us_canada_user_rating[india_us_canada_user_rating.duplicated(['userID', 'bookTitle'])].empty:
    initial_rows = india_us_canada_user_rating.shape[0]

    print('Initial dataframe shape {0}'.format(india_us_canada_user_rating.shape))
    india_us_canada_user_rating = india_us_canada_user_rating.drop_duplicates(['userID', 'bookTitle'])
    current_rows = india_us_canada_user_rating.shape[0]
    print('New dataframe shape {0}'.format(india_us_canada_user_rating.shape))
    print('Removed {0} rows'.format(initial_rows - current_rows))
```

```python
india_us_canada_user_rating_pivot = india_us_canada_user_rating.pivot(index = 'bookTitle', columns = 'userID', values = 'bookRating').fillna(0)
india_us_canada_user_rating_matrix = csr_matrix(india_us_canada_user_rating_pivot.values)
```

```python
model_knn = NearestNeighbors(metric = 'cosine', algorithm = 'brute')
model_knn.fit(india_us_canada_user_rating_matrix)
```

```python
query_index = np.random.choice(india_us_canada_user_rating_pivot.shape[0])
distances, indices = model_knn.kneighbors(india_us_canada_user_rating_pivot.iloc[query_index, :].values.reshape(1, -1), n_neighbors = 6)

for i in range(0, len(distances.flatten())):
    if i == 0:
        print('Recommendations for {0}:\n'.format(india_us_canada_user_rating_pivot.index[query_index]))
    else:
        print('{0}: {1}, with distance of {2}:'.format(i, india_us_canada_user_rating_pivot.index[indices.flatten()[i]], distances.flatten()[i]))
```

```python
india_us_canada_user_rating_pivot2 = india_us_canada_user_rating.pivot(index = 'userID', columns = 'bookTitle', values = 'bookRating').fillna(0)
```

```python
india_us_canada_user_rating_pivot2.head()
```

```python
india_us_canada_user_rating_pivot2.shape
```

```python
X = india_us_canada_user_rating_pivot2.values.T
X.shape
```

```python
SVD = TruncatedSVD(n_components=12, random_state=42)
matrix = SVD.fit_transform(X)
matrix.shape
```

```python
corr = np.corrcoef(matrix)
corr.shape
```

```python
india_us_canada_book_title = india_us_canada_user_rating_pivot2.columns
india_us_canada_book_list = list(india_us_canada_book_title)
b_check= india_us_canada_book_list.index("Animal Farm")
print(b_check)
```

```python
corr_b_check  = corr[b_check]
```
